```python
import pandas as pd

files= [
{"file":"map_cities"},
{"file":"map_cancellation_reasons"},
{"file":"bulk_rides"},
{"file":"map_payment_methods"},
{"file":"map_ride_statuses"},
{"file":"map_vehicle_makes"},
{"file":"map_vehicle_types"}
]

for file in files:
    
    url=f"https://dluberprojectdevmgn.blob.core.windows.net/raw/ingestion/{file['file']}.json?sp=r&st=2026-07-22T20:59:26Z&se=2026-08-01T05:14:26Z&spr=https&sv=2026-02-06&sr=c&sig=RLS%2B6tV0oQGLChqdkS5hOoCDbkt8nVMn2d%2BJgO275Ns%3D"

    df=pd.read_json(url)
    df_spark=spark.createDataFrame(df)

    ## Write Data into Bronze Layer
    df_spark.write.format('delta').mode("overwrite").option("overwriteSchema", "true").saveAsTable(f"db_mgn.bronze.{file['file']}")

```
